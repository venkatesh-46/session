```python
data_old=[[3.694,3.694,3.655,3.687,3.694,3.669,3.695,3.655,3.693,3.69,3.693,3.681,3.684]]
```

```python
data_new=[[4.097,4.099,4.078,4.084,4.107,4.091,4.098,4.084,4.089,4.095,4.099,4.098,4.104]]
```

```python
soc1_mapping_old=pd.DataFrame(data_old).transpose()
```

```python
soc1_mapping_new=pd.DataFrame(data_new).transpose()
```

```python
soc1_mapping_old.index=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13']
```

```python
soc1_mapping_new.index=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13']
```

```python
soc1_mapping_old
```

```python
soc1_mapping_new
```

```python
soc_data=[[33,33,27,32,33,29,33,27,33,33,33,32,32]]
```

```python
soc_data=pd.DataFrame(soc_data).transpose()
```

```python
soc_data.index=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13']
```

```python
soc_data
```

```python
s1_old_soc_mapping=pd.merge(soc1_mapping_old,soc_data,left_index=True,right_index=True)
```

```python
s1_old_soc_mapping
```

```python
s1_old_soc_mapping=s1_old_soc_mapping.rename(columns={'0_x':'Old_Voltage','0_y':'SOC_old'})
s1_old_soc_mapping
```

```python
soc_data_new=[[89,90,86,87,91,88,90,87,88,89,90,90,91]]
```

```python
soc_data_new=pd.DataFrame(soc_data_new).transpose()
```

```python
soc_data_new.index=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13']
```

```python
soc_data_new
```

```python
s1_new_soc_mapping=pd.merge(soc1_mapping_new,soc_data_new,left_index=True,right_index=True)
```

```python
s1_new_soc_mapping
```

```python
s1_new_soc_mapping=s1_new_soc_mapping.rename(columns={'0_x':'new_Voltage','0_y':'SOC_new'})
s1_new_soc_mapping
```

```python
s1_soc=pd.merge(s1_old_soc_mapping,s1_new_soc_mapping,left_index=True,right_index=True)
s1_soc
```

```python
s1_soc['SOC_old']=s1_soc['SOC_old']/100
```

```python
s1_soc['SOC_new']=s1_soc['SOC_new']/100
```

```python
s1_soc
```

```python
s1_soc['CC']=s1_soc['SOC_new']-s1_soc['SOC_old']
```

```python
s1_soc
```

```python
s1_soc['CC']=13*s1_soc['CC']
s1_soc
```

```python
s1_soc['SOC_new-SOC_old']=s1_soc['SOC_new']-s1_soc['SOC_old']
s1_soc
```

```python
s1_soc['constant']=13*s1_soc['SOC_new-SOC_old']
s1_soc
```

```python
s1_soc['SOH']=s1_soc['CC']/s1_soc['constant']
s1_soc
```

```python
s1_soc['SOH in Percent']=s1_soc['SOH']*100
s1_soc
```
